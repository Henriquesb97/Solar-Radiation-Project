```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 8500)

df=[]
df=pd.DataFrame(df)

nome_meses=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']

for y in range (4,10): 
    for m in nome_meses:
        a=pd.read_excel(r'C:\Users\Katia Suely\Desktop\Planilhas do programa\201{}\201{}{}.xlsx'.format(y,y,m))
        df=df.append(a)
```

```python
df.set_index('Unnamed: 0',inplace=True)
```

```python
plt.gcf().set_size_inches(300, 10)
sns.cubehelix_palette(start=.5, rot=-.6, as_cmap=True)
chart=sns.boxplot(x=df.index.strftime('%m--%H:%M'),y=df['SolarRad'], data=df,palette="Spectral")
chart.set_xticklabels(chart.get_xticklabels(), rotation=65)
plt.title('Radiação por horário por mês',loc='left')
plt.ylabel('Radiação solar')
plt.xlabel('Período')
plt.show()
```
