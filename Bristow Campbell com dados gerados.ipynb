```python
def KTBC(Î”t,A,B,C):
    return A*(1-np.exp(-B*Î”t**C))
```

```python
g1 = [5,0.9,3.0]
g2 = [0.73,0.6,2.0]


n1= len(grupo1.KT)
y1= np.empty(n1)

n2= len(grupo2.KT)
y2= np.empty(n2)


for i in range (n1):
    y1[i]=KTBC(grupo1.KT[i],g1[0],g1[1],g1[2])

for i in range (n2):
    y2[i]=KTBC(grupo2.KT[i],g2[0],g2[1],g2[2])   

#plt.scatter(grupo1.Î”t,grupo1.KT,color='r',alpha=0.3)
plt.scatter(grupo2.Î”t,grupo2.KT,color='b',alpha=0.3)
#plt.scatter(grupo1.Î”t,y1,color='g')
plt.scatter(grupo2.Î”t,y2,color='y')

plt.show()
```
