```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
pd.set_option('display.max_rows',40000)
pd.set_option('display.max_columns',2200)
```

```python
grupo2=pd.read_excel(r'C:\Users\Katia Suely\Desktop\GrupoAnual2.xlsx')
grupo2
```

```python
def KTBC(Î”t,A,B,C):
    return A*(1-np.exp(-B*Î”t**C))
```

```python
g = [0.75,0.7,2.4]

%matplotlib inline

n1= len(grupo1.KT)
y1= np.empty(n1)

n2= len(grupo2.KT)
y2= np.empty(n2)


for i in range (n1):
    y1[i]=KTBC(grupo1.KT[i],g[0],g[1],g[2])

for i in range (n2):
    y2[i]=KTBC(grupo2.KT[i],g[0],g[1],g[2])   

plt.scatter(grupo1.Î”t,grupo1.KT,color='r',alpha=0.3)
plt.scatter(grupo2.Î”t,grupo2.KT,color='b',alpha=0.3)
plt.scatter(grupo1.Î”t,y1,color='g')
plt.scatter(grupo2.Î”t,y2,color='y')

plt.show()
```

```python
from scipy.optimize import curve_fit
c1,cov1 = curve_fit(KTBC,grupo1.Î”t.values,grupo1.KT.values,g)
c2,cov2 = curve_fit(KTBC,grupo2.Î”t.values,grupo2.KT.values,g)

for i in range (n1):
    y1[i]=KTBC(grupo1.Î”t[i],c1[0],c1[1],c1[2])
    
for i in range (n2):
    y2[i]=KTBC(grupo2.Î”t[i],c2[0],c2[1],c2[2])
    
plt.scatter(grupo1.Î”t,grupo1.KT,color='r',alpha=0.3)
plt.scatter(grupo2.Î”t,grupo2.KT,color='b',alpha=0.3)
plt.scatter(grupo1.Î”t,y1,color='g')
plt.scatter(grupo2.Î”t,y2,color='y')
plt.show()
```

```python
from sklearn.metrics import r2_score
print('R^2: ', r2_score(grupo1.KT,y1))
print('R^2: ', r2_score(grupo2.KT,y2))
```
