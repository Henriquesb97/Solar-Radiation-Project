# Modelo de Hargreaves & Samani (HS):

```python
Kr1=0.00185*pow(TabelaTempmax-TabelaTempmin,2)
```

```python
Kr2=Kr1-0.0433*(TabelaTempmax-TabelaTempmin)+0.4023
```

```python
KTHS=Kr2*pow(TabelaTempmax-TabelaTempmin,0.5)
```

```python
KTHS
```

```python
pd.DataFrame(KTHS.mean()).T
```
