```python
deltaT.drop(deltaT.index[59],inplace=True)
```

```python
metodo=[]
metodo=pd.DataFrame(metodo)
metodo['Δt']=deltaT.Variação
KT.index=deltaT.Variação.index
metodo['KT']=KT
metodo.index=pd.to_datetime(metodo.index, format='%m-%d')
```

```python
metodo
```

```python
jan=(metodo.index.strftime('%m')=='01')
fev=(metodo.index.strftime('%m')=='02')
mar=(metodo.index.strftime('%m')=='03')
abr=(metodo.index.strftime('%m')=='04')
mai=(metodo.index.strftime('%m')=='05')
jun=(metodo.index.strftime('%m')=='06')
jul=(metodo.index.strftime('%m')=='07')
ago=(metodo.index.strftime('%m')=='08')
sep=(metodo.index.strftime('%m')=='09')
out=(metodo.index.strftime('%m')=='10')
nov=(metodo.index.strftime('%m')=='11')
dez=(metodo.index.strftime('%m')=='12')
filtgrupo1=jan|ago|sep|out|nov|dez
filtgrupo2=fev|mar|abr|mai|jun|jul
```

```python
grupo1=metodo.loc[filtgrupo1]
grupo2=metodo.loc[filtgrupo2]
```
