```python
KTs=pd.read_excel(r'C:\Users\Katia Suely\Desktop\Valores de KT diario.xlsx')
```

```python
KTs.index=KTs['Unnamed: 0']
KTs=KTs.drop(['Unnamed: 0'],axis=1)
```

```python
KTs
```

```python
metodo=[]
metodo=pd.DataFrame(metodo)
metodo['KT']=KTs.KT
deltaT.index=KTs.index
metodo['Δt']=deltaT.Δt
```

```python
filt=(metodo.Δt>15) 
metodo.loc[filt]
metodo.Δt.replace(metodo.Δt.loc[filt],np.nan,inplace=True)
metodo=metodo.dropna()
```

```python
from scipy.stats import iqr
iqr(metodo.Δt)
Q1 = np.quantile(metodo.Δt,0.25)
Q3 = np.quantile(metodo.Δt,0.75)
IQR = Q3 - Q1
IQR
Q3*1.5
```

```python
Q1*1.5
```

```python
filt2=((metodo.Δt>Q1*1.5) & (metodo.Δt<Q3*1.5))
a=metodo.loc[filt2]
a.shape
```

```python
from sklearn.ensemble import IsolationForest
isolation_forest = IsolationForest(n_estimators=100)
isolation_forest.fit(metodo.Δt.values.reshape(-1, 1))
xx = np.linspace(metodo.Δt.min(), metodo.Δt.max(), len(metodo)).reshape(-1,1)
anomaly_score = isolation_forest.decision_function(xx)
outlier = isolation_forest.predict(xx)
plt.figure(figsize=(10,4))
plt.plot(xx, anomaly_score, label='anomaly score')
plt.fill_between(xx.T[0], np.min(anomaly_score), np.max(anomaly_score), 
                 where=outlier==-1, color='r', 
                 alpha=.4, label='outlier region')
plt.legend()
plt.ylabel('anomaly score')
plt.xlabel('Valores de Δt')
plt.show();
```

```python
filt=((metodo.Δt>=4) & (metodo.Δt<=6.5))
a=metodo.loc[filt]
```

```python
a.Δt
```

```python
def KTBC(Δt,A,B,C):
    return A*(1-np.exp(-B*Δt**C))
```

```python
g = [0.75,0.03,2.4]

%matplotlib inline

n= len(a['KT'])
y= np.empty(n)

for i in range (n):
    y[i]=KTBC(a['Δt'][i],g[0],g[1],g[2])

plt.scatter(a['Δt'],a['KT'])
plt.scatter(a['Δt'],y)
plt.show()
```

```python
from scipy.optimize import curve_fit
c,cov = curve_fit(KTBC,a['Δt'].values,a['KT'].values,g)

for i in range (n):
    y[i]=KTBC(a['Δt'][i],c[0],c[1],c[2])
    
plt.scatter(a['Δt'],a['KT'])
plt.scatter(a['Δt'],y)
```

```python
from sklearn.metrics import r2_score
print('R^2: ', r2_score(a['KT'].values,y))
```
